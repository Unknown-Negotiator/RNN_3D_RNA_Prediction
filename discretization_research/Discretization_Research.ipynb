```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.mixture import GaussianMixture

# Assuming your DataFrame is named 'df' and the column of interest is 'gamma'
# Replace these with your actual DataFrame and column names

# Visualize the histogram to confirm the bimodal nature
plt.hist(df['gamma'], bins=30, density=True, alpha=0.5, color='b', label='Histogram')
plt.xlabel('Gamma')
plt.ylabel('Frequency')
plt.title('Histogram of Gamma')
plt.show()

# Reshape the data for GMM
X = df['gamma'].values.reshape(-1, 1)

# Use Gaussian Mixture Modeling to fit the data
gmm = GaussianMixture(n_components=2, random_state=42)
df['cluster'] = gmm.fit_predict(X)

# Separate the data into two DataFrames based on the GMM clusters
cluster_0 = df[df['cluster'] == 0].drop(columns='cluster')
cluster_1 = df[df['cluster'] == 1].drop(columns='cluster')

# Visualize the GMM clusters
plt.scatter(cluster_0.index, cluster_0['gamma'], label='Cluster 0')
plt.scatter(cluster_1.index, cluster_1['gamma'], label='Cluster 1')
plt.xlabel('Index')
plt.ylabel('Gamma')
plt.title('GMM Clusters of Gamma')
plt.legend()
plt.show()
```

```python
def calculate_tertiary_index(angle):
    # Calculate the class index
    if angle < 0:
        angle += 360

    if angle < 93.3:
        class_index = 0
    elif angle < 248.3:
        class_index = 1
    else:
        class_index = 2

    return class_index
```

```python
import os
import json
import matplotlib.pyplot as plt
import pandas as pd

results_folder = 'results'  # Change this to your actual folder path
desired_order = ['SPOT-RNA-1D', 'BinStatsClassClassifier', 'ThreeClassClassifier', '2ClassClassifier', '20ClassClassifier', '30ClassClassifier', 'Regressor']

# Lists to store data for box plots
model_names = []
average_mae_test = []

# Iterate over sub-folders in the results folder
for model_folder in os.listdir(results_folder):
    model_path = os.path.join(results_folder, model_folder)

    if os.path.isdir(model_path) and model_folder in desired_order:
        model_names.append(model_folder)

        # Read MAE scores from json files
        mae_test_path = os.path.join(model_path, 'mae_test.json')

        with open(mae_test_path, 'r') as f:
            mae_test_data = json.load(f)
            average_mae_test.append(sum(mae_test_data.values()) / len(mae_test_data))

# Create a DataFrame for box plots
data = {'Model': model_names, 'Average MAE Test': average_mae_test}
df = pd.DataFrame(data)

# Use Categorical to set the desired order of 'Model' column
df['Model'] = pd.Categorical(df['Model'], categories=desired_order, ordered=True)

# Create box plot for the testing data
fig, ax = plt.subplots(figsize=(8, 6))
df.boxplot(column='Average MAE Test', by='Model', ax=ax)

# Adjust plot labels and title
#ax.set_title('Average MAE Test')
plt.suptitle('Average Test MAE Across Models')

# Rotate x-axis labels
ax.set_xticklabels(ax.get_xticklabels(), rotation=45, ha='right')

# Add mean values above each boxplot
means = df.groupby('Model')['Average MAE Test'].mean().values
for i, mean in enumerate(means):
    plt.text(i, mean + 0.1, f'Mean: {mean:.2f}', ha='center', va='bottom', color='blue') # fontweight='bold'

# Save the plot
plt.savefig('average_mae_plot.png')

# Show the plot
plt.show()
```
